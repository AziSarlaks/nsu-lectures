```python
#1 - семантика
"""Реализуйте функцию find_most_similar(target_word, words_dict, top_k=3), которая находит top_k самых семантически близких слов
к целевому слову на основе их векторных представлений.
Функция должна:
    Использовать косинусную близость
    Исключать целевое слово из результатов
    Возвращать отсортированный список кортежей (слово, схожесть)
    Обрабатывать случай отсутствия целевого слова в словаре

Используйте любой словарь
dct1 = {'кошка': [0.9, 0.1, 0.8, 0.2],
           'собака': [0.8, 0.2, 0.7, 0.3],
           'тигр': [0.7, 0.3, 0.9, 0.1],
           'лев': [0.6, 0.4, 0.8, 0.2],
           'рыба': [0.1, 0.9, 0.2, 0.8], 
           'птица': [0.2, 0.8, 0.3, 0.7] }

dct2 = {'яблоко': [0.9, 0.1, 0.2],
          'груша': [0.8, 0.2, 0.3],
          'банан': [0.7, 0.3, 0.4],
          'апельсин': [0.6, 0.4, 0.1],
          'машина': [0.1, 0.9, 0.8], 
          'дом': [0.2, 0.8, 0.7]}

dct3 = {'красный': [0.9, 0.1, 0.2],
          'оранжевый': [0.8, 0.3, 0.2],
          'желтый': [0.7, 0.5, 0.1],
          'зеленый': [0.1, 0.9, 0.2],
          'синий': [0.2, 0.3, 0.9],
          'фиолетовый': [0.5, 0.2, 0.8]}  
          
Вывод будет выглядеть так:
Самые близкие слова к "ВАШЕ СЛОВО"
1. Первое слово
2. Второе слово
3. ...
"""
```

```python
# Enter your code here
def cosine_sim(a, b):
    if all(x == 0 for x in a) or all(x == 0 for x in b):
        return 0.0
    dp = sum(x * y for x, y in zip(a, b))
    norm_a = sum(x ** 2 for x in a) ** 0.5
    norm_b = sum(x ** 2 for x in b) ** 0.5
    res = (dp / (norm_a * norm_b))
    return res

def find_most_similar(target_word, words_dict, top_k=3):
    target_vector = words_dict[target_word]
    sims = []
    for word, vector in words_dict.items():
        if word != target_word:
            sim = cosine_sim(target_vector, vector)
            sims.append((word, sim))
    
    sims.sort(key=lambda x: x[1], reverse=True)
    return sims[:top_k]

dict1 = {'кошка': [0.9, 0.1, 0.8, 0.2],
        'собака': [0.8, 0.2, 0.7, 0.3],
        'тигр': [0.7, 0.3, 0.9, 0.1],
        'лев': [0.6, 0.4, 0.8, 0.2],
        'рыба': [0.1, 0.9, 0.2, 0.8], 
        'птица': [0.2, 0.8, 0.3, 0.7]}

target = 'кошка'
result = find_most_similar(target, dict1, 6)

print(f'Самые близкие слова к "{target}":')
for i, (word, sim) in enumerate(result, 1):
    print(f'{i}. {word}')
```

```python
# Enter your code here
import sqlite3

def borrow_book(book_id):
    conn = sqlite3.connect('seminar10.db')
    cursor = conn.cursor()
    
    cursor.execute('UPDATE books SET is_available = FALSE WHERE id = ?', (book_id,))
    conn.commit()
    
    conn.close()

def return_book(book_id):
    conn = sqlite3.connect('seminar10.db')
    cursor = conn.cursor()
    
    cursor.execute('UPDATE books SET is_available = TRUE WHERE id = ?', (book_id,))
    conn.commit()
    
    conn.close()

return_book(1)
```
