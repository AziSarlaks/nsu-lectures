```python
#36
"""Юный инженер! Твой робот-исследователь только что вернулся из миссии по неизведанной территории. 
Он записал координаты всех точек своего маршрута. Тебе нужно рассчитать общее расстояние, которое преодолел робот!
На вход программе подается кортеж координат (x, y) точек маршрута - trajectory = ((0, 0), (3, 0), (3, 4), (0, 0))
Вычислить суммарную длину пути между всеми точками.
Ожидаемый результат: 12.0 """
```

```python
def calculate_total_distance(points):
    a = 0
    b = 0
    s = 0
    for i in points:
        if i[0] == a or i[1] == b:
            s += abs(a - i[0])
            s += abs(b - i[1])
        else:
            s += (abs(i[0]-a)**2 + abs(i[1]-b)**2)**0.5 
        a = i[0]
        b = i[1]
    return s

# Тест
trajectory = ((0, 0), (3, 0), (3, 4), (0, 0))
print(calculate_total_distance(trajectory))
```

```python
# Напишите Ваш код здесь
def is_correct_bracket(text):
    o = 0
    for i in text:
        if i == "(":
            o+=1
        if i == ")":
            o-=1
            if o < 0: return False
    return True if o == 0 else False
print(is_correct_bracket("())((((())))"))
```

```python
# Напишите Ваш код здесь

def number_to_words(num):
    digits = {
        0: '', 1: 'один', 2: 'два', 3: 'три', 4: 'четыре', 
        5: 'пять', 6: 'шесть', 7: 'семь', 8: 'восемь', 9: 'девять'
    }
    ten = {
        10: 'десять', 11: 'одиннадцать', 12: 'двенадцать', 13: 'тринадцать',
        14: 'четырнадцать', 15: 'пятнадцать', 16: 'шестнадцать', 
        17: 'семнадцать', 18: 'восемнадцать', 19: 'девятнадцать'
    }
    tens = {
        2: 'двадцать', 3: 'тридцать', 4: 'сорок', 5: 'пятьдесят',
        6: 'шестьдесят', 7: 'семьдесят', 8: 'восемьдесят', 9: 'девяносто'
    }
    
    # Проверка на особые случаи (10-19)
    if 10 <= num <= 19:
        return ten[num]
    
    tens_digit = num // 10
    units_digit = num % 10
    
    if tens_digit == 0:
        return digits[units_digit]
    elif units_digit == 0:
        return tens[tens_digit]
    else:
        return tens[tens_digit] + ' ' + digits[units_digit]
print(number_to_words(99))
```

```python
# Напишите Ваш код здесь
n = int(input())

x = []
for i in range(n):
    row = list(map(int, input().split()))
    x.append(row)

upper_sum = 0
right_sum = 0
lower_sum = 0
left_sum = 0

for i in range(n):
    for j in range(n):
        if i < j and i + j < n - 1: upper_sum += x[i][j]
        elif i < j and i + j > n - 1: right_sum += x[i][j]
        elif i > j and i + j > n - 1: lower_sum += x[i][j]
        elif i > j and i + j < n - 1: left_sum += x[i][j]
print(f"Верхняя четверть: {upper_sum}")
print(f"Правая четверть: {right_sum}")
print(f"Нижняя четверть: {lower_sum}")
print(f"Левая четверть: {left_sum}")
```
