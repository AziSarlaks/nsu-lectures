```python
#21
"""Напишите функцию, которая принимает список любой глубины вложенности и возвращает новый "плоский" список,
содержащий все элементы исходного списка в том же порядке, но без вложенности. 

Input1:
[1, [2, [3, [4, 5]], 6], 7]
Output1:
[1, 2, 3, 4, 5, 6, 7]

Input2:
[1, 2, 3, 4, 5]
Output2:
[1, 2, 3, 4, 5]

Input3:
[[[[1]]]]
Output3:
[1]

Input4:
[1, [2], [3, [4, [5]]]]
Output4:
[1, 2, 3, 4, 5]

Input5:
[]
Output5:
[]
                                                 """
```

```python
# Напишите Ваш код здесь
def fl(l):
    result = []
    
    for i in l:
        if isinstance(i, list):
            result.extend(fl(i))
        else:
            result.append(i)
    
    return result

print(fl([1, [2], [3, [4, [5]]]]))
```

```python
# Напишите Ваш код здесь
import csv
import matplotlib.pyplot as plt

data = [
    ['voltage', 'current'],
    [0.0, 0.00],
    [0.5, 0.12],
    [1.0, 0.25],
    [1.5, 0.38],
    [2.0, 0.50],
    [2.5, 0.63],
    [3.0, 0.75],
    [3.5, 0.88],
    [4.0, 1.00],
    [4.5, 1.13],
    [5.0, 1.25],
    [5.5, 1.38],
    [6.0, 1.50],
    [6.5, 1.63],
    [7.0, 1.75],
    [7.5, 1.88],
    [8.0, 2.00]
]
    
with open('voltage_current.csv', 'w') as file:
    writer = csv.writer(file)
    writer.writerows(data)

    voltages = []
    currents = []
    
with open('voltage_current.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)
    
    for row in reader:
        voltages.append(float(row[0]))
        currents.append(float(row[1]))


plt.figure(figsize=(10, 6))
plt.plot(voltages, currents, 'bo-', linewidth=2, markersize=10)
plt.title('Вольт-амперная характеристика', fontsize=14)
plt.xlabel('Напряжение, В', fontsize=12)
plt.ylabel('Ток, А', fontsize=12)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
