```python
# 1 
"""Вам доступна функция traffic(), реализованная с помощью цикла while, которая принимает в качестве аргумента число 
n и выводит n раз строку Не парковаться. Перепишите данную функцию с использованием рекурсии, чтобы она выполняла ту же задачу.
Input 1:
traffic(3)
Output 1:
Не парковаться
Не парковаться
Не парковаться

Input 2:
traffic(5)
Output 2:
Не парковаться
Не парковаться
Не парковаться
Не парковаться
Не парковаться

Input 3:
traffic(0)
Output 3:
"""
```

```python
# Перепишите код
def traffic(n):
    if n == 1:
        print("Не парковаться")
        return;
    print("Не парковаться")
    traffic(n-1)
traffic(1)
```

```python
#3
"""Реализуйте функцию triangle() с использованием рекурсии, которая принимает один аргумент:
h — натуральное число
Input 1:
triangle(3)
Output 1:
***
**
*

Input 2:
triangle(5)
Output 2:
*****
****
***
**
*

 """
```

```python
# Напишите Ваш код здесь
def triangle(h):
    if h == 1:
        print("*"*h)
        return;
    print("*"*h)
    triangle(h-1)
triangle(3)
triangle(5)
```

```python
#4
"""Напишите программу с использованием рекурсивной функции, которая выводит изображение песочных часов, составленное из цифр 
1,2,3, и 4:

1111111111111111    # 16 раз
  222222222222      # 12 раз
    33333333        # 8 раз
      4444          # 4 раза
    33333333        # 8 раз
  222222222222      # 12 раз
1111111111111111    # 16 раз
                  
Формат входных данных
На вход программе ничего не подается. """
```

```python
# Напишите Ваш код здесь
def sample(n=16, p=20):
    if n == 4:
        print(" "*6 + "4"*4)
        sample(n+4, n)
    elif n == 8:
        print(" "*4 + "3"*8)
        if p > n:
            sample(n-4, n)
        else:
            sample(n+4, n)
    elif n == 12:
        print(" "*2 + "2"*12)
        if p > n:
            sample(n-4, n)
        else:
            sample(n+4, n)
    elif n == 16:
        print("1"*16)
        if p < n:
            return;
        else:
            sample(n-4, n)
sample()
```

```python
#5
"""Реализуйте функцию print_digits() с использованием рекурсии, которая принимает один аргумент:
number — натуральное число
Функция должна выводить все цифры числа number, начиная со старших разрядов, каждое на отдельной строке.
Input 1:
print_digits(12345)
Output 1:
1
2
3
4
5

Input 2:
print_digits(2077)
Output 2:
2
0
7
7

Input 3:
print_digits(8)
Output 3:
8                  """
```

```python
# Напишите Ваш код здесь
def print_digits(num):
    if len(str(num)) == 1:
        print(str(num)[0])
        return;
    print(str(num)[0])
    print_digits(str(num)[1:])
print_digits(12345)
```

```python
#6
"""Напишите программу с использованием рекурсии, которая принимает на вход число и выводит количество цифр в этом числе.
Формат входных данных
На вход программе подается неотрицательное целое число.
Формат выходных данных
Программа должна определить количество цифр в введенном числе и вывести полученный результат.
Input 1:
50
Output 1:
2

Input 2:
17488
Output 2:
5

Input 3:
7
Output 3:
1    """
```

```python
# Напишите Ваш код здесь
def CoD(n):
    if n < 10:
        return 1
    return 1 + CoD(n//10)
print(CoD(17488))
```

```python
#7
"""Реализуйте функцию get_pow() с использованием рекурсии, которая принимает два аргумента в следующем порядке:
a — положительное целое число
n — неотрицательное целое число
Функция должна вычислять значение a в степени n и возвращать полученный результат.
При решении не используйте оператор возведения в степень **.
Input 1:
print(get_pow(5, 2))
Output 1:
25

Input 2:
print(get_pow(99, 0))
Output 2:
1

Input 3:
print(get_pow(2, 10))
Output 3:
1024       """
```

```python
# Напишите Ваш код здесь
def get_pow(a, n):
    if n == 0:
        return 1
    if n % 2 == 0:
        return get_pow(a*a, n/2)
    if n % 2 != 0:
        return a*get_pow(a, n-1)
print(get_pow(2, 10))
```

```python
#8
"""Реализуйте функцию recursive_sum() с использованием рекурсии, которая принимает два аргумента в следующем порядке:
a — неотрицательное целое число
b — неотрицательное целое число
Функция должна возвращать сумму чисел a и b. При вычислении суммы функция:
не должна использовать циклы
из всех арифметических операций должна использовать только +1 и −1

Input 1:
print(recursive_sum(10, 22))
Output 1:
32

Input 2:
print(recursive_sum(99, 0))
Output 2:
99

Input 3:
print(recursive_sum(0, 0))
Output 3:
0                               """
```

```python
# Напишите Ваш код здесь
def recursive_sum(a, b):
    if b == 0: return a
    return recursive_sum(a + 1, b - 1)

print(recursive_sum(10, 22))
print(recursive_sum(99, 0))
print(recursive_sum(0, 0))
```

```python
#9
"""Реализуйте функцию to_binary() с использованием рекурсии, которая принимает один аргумент:
number — неотрицательное целое число
Функция должна возвращать строковое представление числа number в двоичной системе счисления.
Input 1:
print(to_binary(15))
Output 1:
1111

Input 2:
print(to_binary(0))
Output 2:
0

Input 3:
print(to_binary(1))
Output 3:
1                """
```

```python
# Напишите Ваш код здесь
def to_binary(number):
    if number <= 1: return str(number)
    return to_binary(number // 2) + str(number % 2)
print(to_binary(15))
print(to_binary(0))
print(to_binary(1))
```

```python
#10
"""В первый год в пруду живет 77 лягушек. Каждый год из пруда вылавливают 30 лягушек, а оставшиеся размножаются,
и их становится в три раза больше. Так, количество лягушек k-й год  может быть описано формулой:
Fk = 3*(F(k-1) - 30)
Реализуйте функцию number_of_frogs() с использованием рекурсии, которая принимает один аргумент:
year — натуральное число
Функция должна возвращать единственное число — количество лягушек в пруду в году year.
Input 1:
print(number_of_frogs(2))
Output 1:
141

Input 2:
print(number_of_frogs(10))
Output 2:
629901

Input 3:
print(number_of_frogs(1))
Output 3:
77        """
```

```python
# Напишите Ваш код здесь
def number_of_frogs(y):
    if y == 1:
        return 77
    return 3*(number_of_frogs(y-1) - 30)

print(number_of_frogs(2))
print(number_of_frogs(10))
print(number_of_frogs(1))
```
