```python
#1-45 (2 балла)
"""Реализуйте декоратор prefix, который принимает два аргумента в следующем порядке:
string — произвольная строка
to_the_end — булево значение, по умолчанию равное False
Декоратор должен добавлять строку string к возвращаемому значению декорируемой функции. 
Если to_the_end имеет значение True, строка string добавляется в конец, если False — в начало.
Также декоратор должен сохранять имя и строку документации декорируемой функции.
Гарантируется, что возвращаемым значением декорируемой функции является объект типа str.
Не забывайте про то, что декоратор не должен поглощать возвращаемое значение декорируемой функции, 
а также должен уметь декорировать функции с произвольным количеством позиционных и именованных аргументов.
Input 1:
@prefix('€')
def get_bonus():
    return '2000'   
print(get_bonus())
Output 1:
€2000

Input 2:
@prefix('$$$', to_the_end=True)
def get_bonus():
    return '2000'
print(get_bonus())
Output 2:
2000$$$    """
```

```python
# Enter your code here
import functools

def prefix(string, to_the_end=False):
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*a, **kwa):
            result = func(*a, **kwa)
            if to_the_end:
                return result + string
            else:
                return string + result
        return wrapper
    return decorator

@prefix('€')
def get_bonus():
    """DOC"""
    return '2000'   
print(get_bonus())
print(get_bonus.__name__)
print(get_bonus.__doc__)

@prefix('$$$', to_the_end=True)
def get_bonus():
    return '2000'
print(get_bonus())
```

```python
#4-45 (2 балла)
"""Саша создаёт программу для водителей. 
В ней вы указываете расстояние, которое планируете проехать, расход топлива на 100км вашей машины, цену на бензин в вашем регионе.
Программа выводит на экран информацию о будущих финансовых расходах. Саша вероломно использует статический метод для подсчёта.

Задание:
Создайте класс Driver
Объявите статический метод calculate_fuel_costs() с параметрами (distance, fuel, price). Напоминашка: @staticmethod.
- Создайте переменную result, в ней будет хранится результат подсчётов.
- Готовая формула подсчётов price * (fuel / 100) * distance.
- Выведите на экран округлённое значение result, используя функцию round(). 
Значение округлите до сотых. Постарайтесь сделать округление имено с round(), иначе ответы могут не совпасть.
Оставшаяся часть кода уже написана, вам нужно сделать то, что указано в задании."""
```

```python
# Enter your code here
class Driver:
    @staticmethod
    def calculate_fuel_costs(distance, fuel, price):
        result = price * (fuel / 100) * distance
        print(round(result, 2))

# код ниже пожалуйста не удаляйте
sasha = Driver()
sasha.calculate_fuel_costs(3, 7, 50)
sasha.calculate_fuel_costs(100, 7, 50)
sasha.calculate_fuel_costs(50, 7, 50)
```

```python
#5-45 (5 баллов)
"""Напишите программу, которая удаляет все книги, изданные раньше 2000 года. 
Перед удалением выведите список этих книг на экран для подтверждения действия.
Пример вывода:
Следующие книги будут удалены:
ID: 1, Война и мир - Лев Толстой (1869г.)
ID: 2, Преступление и наказание - Федор Достоевский (1866г.)
ID: 3, 1984 - Джордж Оруэлл (1949г.)
ID: 4, Мастер и Маргарита - Михаил Булгаков (1967г.)
ID: 5, Гарри Поттер - Джоан Роулинг (1997г.)

Подтвердите удаление (y/n):  y
Удалено 5 книг"""
```

```python
# Enter your code here
import sqlite3
conn = sqlite3.connect('seminar10.db')
cursor = conn.cursor()
cursor.execute("SELECT id, title, author, year FROM books WHERE year < 2000")
books_to_delete = cursor.fetchall()
if books_to_delete:
    print("Следующие книги будут удалены:")
    for book in books_to_delete:
        print(f"ID: {book[0]}, {book[1]} - {book[2]} ({book[3]}г.)")
    conf = input("\nПодтвердите удаление (y/n):  ")
    if conf.lower() == 'y':
        cursor.execute("DELETE FROM books WHERE year < 2000")
        conn.commit()
        print(f"Удалено {len(books_to_delete)} книг")
conn.close()
```
