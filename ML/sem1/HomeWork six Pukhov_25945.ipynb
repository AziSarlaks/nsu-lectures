```python
"""Разработайте программу для ведения списка покупок. Система должна:
1. Инициализация (5 баллов)
Создать словарь покупок shopping_list:
молоко: 1
хлеб: 2
яйца: 10

2. Добавление товаров с обработкой ошибок (10 баллов)
Функция add_prod(товар, количество):
    Добавляет товар в список
    Обрабатывает ошибки:
        Отрицательное количество
        Нечисловое количество

3. Рекурсивный поиск товаров (10 баллов)
Рекурсивная функция find_prod(количество):
Ищет товары, которых нужно купить больше указанного количества
Возвращает список найденных товаров

4. Расчет стоимости (5 баллов)
Рассчитать общую стоимость на основе:
    Списка цен: [('молоко', 80), ('хлеб', 40), ('яйца', 10)]
    Вывести общую сумму
"""
```

```python
# Enter your code here
shopping_list = None
def init(values=['молоко', 'хлеб', 'яйца'], count=[1, 2, 10]):
    global shopping_list
    shopping_list = dict(zip(values, count))

def add_prod(value, count):
    global shopping_list
    try:
        if count < 0:
            raise ValueError
        shopping_list[value] += count
    except KeyError:
        shopping_list[value] = count
    except ValueError:
        print("Введено нечисловое или отрицательное значение")
    except TypeError:
        print("Введено нечисловое или отрицательное значение")

def find_prod(count, values=None, result=[]):
    if values == None: values = list(shopping_list.items())
    if not values: return result
    v, c = values[0]
    if c > count:
        result.append(v)
    return find_prod(count, values[1:], result)
    
def calc_prices(prices=[('молоко', 80), ('хлеб', 40), ('яйца', 10)]):
    global shopping_list
    total_cost = 0
    price_dict = dict(prices)
    for value, count in shopping_list.items():
        try:
            total_cost += price_dict[value] * count
        except KeyError:
            print(f"Товар {value} не найден")
    
    return total_cost


init()
print(shopping_list)

add_prod('Яблоко', 100)
print(shopping_list)

print(find_prod(2))

print(calc_prices())
```
